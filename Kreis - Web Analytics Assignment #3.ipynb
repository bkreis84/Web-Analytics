```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# importing using a dataframe
csv = "https://raw.githubusercontent.com/bkreis84/Web-Analytics/master/Consumer_Complaints.csv"
df = pd.read_csv(csv)
```

```python
#Kind of misusing my data here but I'll look at the types of complaints that companies had
graph = nx.from_pandas_dataframe(df, 'Product', 'Company', create_using=nx.Graph())
```

```python
# draw
nx.draw_spring(graph, with_labels=False, node_size=50)
```
