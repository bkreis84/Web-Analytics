```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create the empty graph
SN = nx.Graph()
#Add nodes that represent individuals
SN.add_nodes_from(['Rosie','Brian', 'Paul', 'Steph', 'Eli', 'Clark', 'Pepe'])
nx.draw(SN, node_size = 900, with_labels = True)

```

```python
# Add the connections
SN.add_edges_from([('Brian','Steph'), ('Steph', 'Rosie'), ('Steph', 'Pepe'), ('Steph', 'Paul'), ('Steph', 'Clark'), ('Brian', 'Steph'), ('Brian', 'Eli')])
nx.draw(SN, node_size = 900, with_labels = True)
#The following represents which pets my wife and I have brought into the house since we have been together. 
```

```python
#Show the connections of Steph
SN.neighbors('Steph')
```

```python
SN.remove_nodes_from(['Rosie', 'Pepe'])
nx.draw(SN, node_size = 900, with_labels = True)
# We lost a couple pets
```

```python
SN.add_edges_from([('Brian','Charlie'), ('Steph', 'Charlie')])
nx.draw(SN, node_size = 900, with_labels = True)
# But added a son
```
